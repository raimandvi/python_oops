## Abstraction in Python

Abstraction focuses on hiding implementation details and showing only essential features to the user.

####Why Abstraction is needed

*   Improves code readability
*   Reduces complexity


*   Provides security





**1. Abstract Base Class (ABC)**



*   Cannot create object of abstract class.



```python
from abc import ABC, abstractmethod

class Shape(ABC):
  @abstractmethod
  def area(self):
    pass
```

**2. Implement Abstract Method**

```python
class Rectangle(Shape):
    def area(self):
        print("Area of Rectangle")

class Circle(Shape):
    def area(self):
        print("Area of Circle")

r = Rectangle()
c = Circle()

r.area()
c.area()
```

**3. Abstract class with constructor**

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def start(self):
        print(self.name, "started")

car = Car("Honda")
car.start()
```
