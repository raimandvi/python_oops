## Polymorphism in Python

Polymorphism means "many forms". It allows the same function or method name to behave differently depending on the object.

**1. Function Polymorphism**



*   Same function (len) works differently.



```python
print(len("Python"))
print(len([1,2,3,4,5]))
```

**2. Method Overriding (Runtime Polymorphism)**

```python
class Animal:
    def sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

animals = [Dog(), Cat()]

for animal in animals:
    animal.sound()
```

**3. Polymorphism with Different Classes**



*   Same method name
*   Different behaviour


```python
class Bird:
    def fly(self):
        print("Bird can fly")

class Airplane:
    def fly(self):
        print("Airplane can fly")

def flying_test(obj):
    obj.fly()

b = Bird()
a = Airplane()

flying_test(b)
flying_test(a)
```

**4. Polymorphism Using Inheritance**

```python
class Shape:
    def area(self):
        print("Calculating area")

class Square(Shape):
    def area(self):
        print("Area of square")

class Circle(Shape):
    def area(self):
        print("Area of circle")

s = Square()
c = Circle()

s.area()
c.area()
```

**5. Operator Overloading**



*   Same operator, different behaviour



```python
print(10 + 20)
print("Hello " + "World")
```
